1) Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica. 

```python
def notas_calculo(diccionario):
    notas = pd.Series(diccionario)
    resumen = pd.Series({
        'Nota mínima': np.round(notas.min(),decimals=2),
        'Nota máxima': np.round(notas.max(),decimals=2),
        'Nota media': np.round(notas.mean(),decimals=2),
        'Desviación típica': np.round(notas.std(),decimals=2)
    })    
    return resumen.to_string(dtype=False)
```

```python
notas = {'Ana': 4.5,'Luis': 3.0,'Carlos': 2.2,'Sara': 3.1,'Angie': 5}
estadisticas = notas_calculo(notas)
print(estadisticas)
```

2)	Cálculo del Índice de Masa Corporal (IMC) con Pandas.

```python
datos=pd.read_csv("datos.csv",sep=';')
display(datos)
```

```python
datos=datos.drop(['Unnamed: 0'],axis=1)
datos_nuevos = datos.rename(columns={
    'Weight': 'Height',
    'Height': 'Weight'})
display(datos_nuevos)
```

```python
pesos=np.array(datos_nuevos['Weight'])
alturas=np.array(datos_nuevos['Height']/100)
bmi=np.round((pesos/(alturas**2)),decimals=1)
categ = np.empty(len(bmi), dtype=object)
for i in np.arange(len(categ)):
    if bmi[i]<18.5:
        categ[i]='Bajo Peso'
    if 18.5<bmi[i]<24.9:
        categ[i]='Normal'
    if 25<bmi[i]<29.9:
        categ[i]='Sobrepeso'
    if bmi[i]>30:
        categ[i]='Obesidad'
datos_nuevos['BMI']=bmi
datos_nuevos['Categoría']=categ
display(datos_nuevos)
```
